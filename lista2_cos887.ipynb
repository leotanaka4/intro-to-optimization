```python
import matplotlib.pyplot as plt

# Define os pontos
x = [1, 0.52]
y = [1, -0.08]

# Plota os pontos
plt.figure(figsize=(20, 12))  # Define o tamanho do grfico
plt.plot(x, y, marker='o', label='Trajetria')  # Plota a trajetria

# Adiciona um vetor
plt.arrow(x[0], y[0], x[1]-x[0], y[1]-y[0], color='r', width=0.02, head_width=0.1, head_length=0.1)

# Adiciona rtulos aos pontos
plt.text(x[0], y[0], ' (1, 1)', fontsize=16, verticalalignment='bottom')
plt.text(x[1], y[1], ' (0.52, -0.08)', fontsize=16, verticalalignment='bottom')

# Define os limites dos eixos
plt.xlim(0, 1.5)
plt.ylim(-0.5, 1.5)

# Adiciona grade ao grfico
plt.grid(True)

# Adiciona rtulos aos eixos
plt.xlabel('$x_1$', fontsize=18)
plt.ylabel('$x_2$', fontsize=18)

# Adiciona a legenda
plt.legend(fontsize=14)

# Adiciona ttulo
plt.title('Figura 1: Grfico da primeira interao no $R^2$', fontsize=20)

# Exibe o grfico
plt.show()
```
